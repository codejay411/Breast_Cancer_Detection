```python
# split
x=df.iloc[:,2:].values
y=df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# feature scalling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

```python
# craete a function for models
def models(x_train, y_train):
    #logistic regression
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression(random_state=0)
    log.fit(x_train, y_train)
    
    #decision tree
    from sklearn.tree import DecisionTreeClassifier
    tree=DecisionTreeClassifier(criterion='entropy', random_state=0)
    tree.fit(x_train, y_train)
    
    #Random forest classifier
    from sklearn.ensemble import RandomForestClassifier
    forest=RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
    forest.fit(x_train, y_train)
    
    # model accuracy
    print("[0]logistic regration training accuracy : ", log.score(x_train, y_train))
    print("[0]logistic regration training accuracy : ", tree.score(x_train, y_train))
    print("[0]logistic regration training accuracy : ", forest.score(x_train, y_train))
    
    return log, tree, forest
```

```python
model= models(x_train, y_train)
```

```python
# test data accuracy 
from sklearn.metrics import confusion_matrix

for i  in range(len(model)):
    print('model', i)
    cm=confusion_matrix(y_test, model[i].predict(x_test))
    TP = cm[0][0]
    TN = cm[1][1]
    FN = cm[1][0]
    FP = cm[0][1]

    print(cm)
    print('testing accuracy - ', (TP+TN)/(TP+TN+FN+FP))
```

```python
# show another way tp get matrics of the models
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for i  in range(len(model)):
    print('model', i)
    print(classification_report(y_test, model[i].predict(x_test)))
    print(accuracy_score(y_test, model[i].predict(x_test)))
    print()
```
